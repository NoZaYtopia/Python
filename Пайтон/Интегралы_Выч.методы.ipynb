```python
import math as m

from scipy.integrate import solve_ivp
#x_global = x 

def f_f(x,y):
    return([y[1],x*m.exp(-x)-5*y[1]-2*y[0]])
x0 = 0 
y0 = [0, 0] 

sol = solve_ivp(f_f, [x0, 2], y0, method='RK23', rtol=0.0001)
x_f = sol.t
y_f = sol.y[0]
dy_f = sol.y[1]
```
