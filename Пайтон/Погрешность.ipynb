```python
import math
import matplotlib.pyplot as plt
import numpy 
x=[]
f=[]
df=[]
dfh=[]
dfh1=[]
dfh2=[]
r=[]
r1=[]
r2=[]
f_max=math.acos(0.1)
ddf_max=0.1/(1-0.1**2)**1.5

h=2*(((f_max*2**(-24))/ddf_max)**0.5)
h1=(((f_max*2**(-24))/ddf_max)**0.5)
h2=4*(((f_max*2**(-24))/ddf_max)**0.5)

for i in range (0,101,1):
    x.append(numpy.float32(i/500-0.1))


for i in range (0,101,1):
    f.append(numpy.float32(math.acos(x[i])))
    df.append(numpy.float32(-1/(1-x[i]**2)**0.5))
    dfh.append(numpy.float32((math.acos(x[i])-math.acos(x[i]-h))/h))
    dfh1.append(numpy.float32((math.acos(x[i])-math.acos(x[i]-h1))/h1))
    dfh2.append(numpy.float32((math.acos(x[i])-math.acos(x[i]-h2))/h2))
for i in range (0,101,1):
    r.append(numpy.float32(df[i]-dfh[i]))
    r1.append(numpy.float32(df[i]-dfh1[i]))
    r2.append(numpy.float32(df[i]-dfh2[i]))
plt.plot(x,f)
plt.show()
```
