```python
import numpy as np
def ap(x,y,st):
    ap1=[[0]*st for i in range(st)]
    ap2=[0]*st
    ap3=[0]*st
    for i in range (0,st):
        for j in range (0,st):
            for k in range (0,len(x)):
                ap1[i][j]+=x[k]**(i+j)
    for i in range (0,st):
        for k in range (0,len(x)):
            ap3[i]+=y[k]*(x[k])**i
    ap2=np.linalg.inv (ap1). dot (ap3)
    return(ap2)
    
def F_ap(x,st,c):
    temp=0
    for i in range (0,st):
        temp+=c[i]*x**i
    return(temp)

def step_ap(x,y,st):
    z=[]
    c=ap(x,y,st)
    for i in range(0,len(x)):
        z.append(F_ap(x[i],st,c))
    return(z)
```
