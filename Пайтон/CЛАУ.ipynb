```python
def Gaus(L,P):
    M=[[0]*len(L) for i in range(0,len(L))]
    for i in range(0,len(L)):
        M[i]=L[i].copy()
    F=P.copy()
    X=[0]*len(M)
    for k in range(0,len(M)):
        for i in range(k+1,len(M)):
            temp=M[i][k]/M[k][k]
            for j in range(0,len(M)):
                M[i][j]-=M[k][j]*temp
            F[i]-=F[k]*temp
    
    X[len(X)-1]=F[len(X)-1]/M[len(X)-1][len(X)-1]

    
    for i in range(len(X)-2,-1,-1):
        for j in range(len(X)-1,i,-1):
            F[i]-=X[j]*M[i][j]
        X[i]=F[i]/M[i][i]
    return(X)
```

```python
X=Gaus(M,F)
print(X)
#for i in range(0,n):
#print(M[i])
res=0
for i in range(0,len(M)):
    res+=X[i]
print()
print(res)
```

```python
def seidel(A, b, eps):
    n = len(A)
    x = np.zeros(n)  # zero vector
    temp=np.array(A)
    converge = False
    while not converge:
        x_new = np.copy(x)
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

#        converge = np.linalg.norm(x_new - x) <= eps
        converge = np.linalg.norm((temp).dot(x_new)-b) <= eps
        x = x_new
    return(x)
```
