```python
import csv
data=[[0]*13 for x in range(154)]
with open('12_Пермь.csv',newline='') as File:
    file_reader=csv.reader(File, delimiter= ",")
    count=0
    for row in file_reader:
        if (count!=0):
            for i in range (0,13):
                data[count-1][i]=float(row[i])
        count+=1
for i in range (0,154):
    print(data[i])
```

```python
def l(i,first,long,x):
    temp=1
    for j in range (first,long):
        if (i!=j):
#                print(data[j][0],' | ',data[i][0])
            if (data[j][1]==999.9) or (data[i][1]==999.9):
#                    print('break')
                break
            else:
#                print(temp)
                temp=temp*(x-data[j][0])/(data[i][0]-data[j][0])
#                print(x, ' - ',data[j][0],' / ', data[i][0], ' - ',data[j][0])
    return(temp)
```

```python
for i in range(17,12):
    print(l(i,17,12,1883))
```

```python
def L(first,long,x):
    temp=0
    for i in range(first,long):
        if (data[i][1]==999.9):
            break
        temp+=data[i][1]*l(i,first,long,x)
    return(temp)
```

```python
def ap(x):
    ap1=[[0]*x for i in range(x)]
    ap2=[0]*x
    ap3=[0]*x
    for i in range (0,x):
        for j in range (0,x):
            for k in range (0,153):
                if (data[k][1]!=999.9):
                    ap1[i][j]+=(data[k][0]-1865)**(i+j)
    for i in range (0,x):
        for k in range (0,153):
            if (data[k][1]!=999.9):
                ap3[i]+=data[k][1]*(data[k][0]-1865)**i
    ap2=np.linalg.inv (ap1). dot (ap3)
    return(ap2)
```

```python
l=ap(4)
print(l)
```

```python
def F_ap(x,t):
    temp=0
    for i in range (0,t):
        temp+=l[i]*x**i
    return(temp)
```
