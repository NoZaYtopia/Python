```python
import matplotlib.pyplot as plt
import numpy
def down (l,d,r):
    x = numpy.linspace (0, l+0.1, 100)
    y = numpy.linspace (0, l+0.1, 100)
    xgrid, ygrid = numpy.meshgrid(x, y)
    z =d* xgrid * ygrid / (xgrid * ygrid)+r
    return xgrid, ygrid, z
def bot (l,d,r):
    x = numpy.linspace (0, l+0.1, 100)
    z = numpy.linspace (r, r+l+0.1, 100)
    xgrid, zgrid = numpy.meshgrid(x, z)
    y =d* xgrid * zgrid / (xgrid * zgrid)
    return xgrid, y, zgrid
def mid (l,d,r):
    y = numpy.linspace (0, l+0.1, 100)
    z = numpy.linspace (r, r+l+0.1, 100)
    ygrid, zgrid = numpy.meshgrid(y, z)
    x =d* ygrid * zgrid / (ygrid * zgrid)
    return x, ygrid, zgrid
fig = plt.figure()
axes = fig.add_subplot(projection='3d')

x, y, z = down(1,1,12)
axes.plot_surface(x, y, z,color="green", alpha=1)
x, y, z = down(1,0,12)
axes.plot_surface(x, y, z,color="green", alpha=1)
x, y, z = bot(1,1,12)
axes.plot_surface(x, y, z,color="green", alpha=1)
x, y, z = bot(1,0,12)
axes.plot_surface(x, y, z,color="green", alpha=1)
x, y, z = mid(1,1,12)
axes.plot_surface(x, y, z,color="green", alpha=1)
x, y, z = mid(1,0,12)
axes.plot_surface(x, y, z,color="green", alpha=1)

x,y,z= bot(3,3,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)
x,y,z= bot(3,0,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)
x,y,z= mid(3,3,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)
x,y,z= mid(3,0,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)
x,y,z= down(3,3,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)
x,y,z= down(3,0,9)
axes.plot_surface(x, y, z,color="yellow", alpha=1)

x,y,z= bot(9,9,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)
x,y,z= bot(9,0,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)
x,y,z= mid(9,9,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)
x,y,z= mid(9,0,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)
x,y,z= down(9,9,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)
x,y,z= down(9,0,0)
axes.plot_surface(x, y, z,color="red", alpha=0.5)

plt.show()
```
