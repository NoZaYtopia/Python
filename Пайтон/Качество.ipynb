```python
seed=633466621
a=29
b=97
M=1000000
chislo=[0]*16
def RNG_unlimited():
    global seed
    seed =(a*seed+b)%M
    temp=seed
    chislo=[int(a) for a in str(seed)]
    for i in range (0,16-len(chislo)):
            chislo.insert(0,0) 
    temp_mas1=chislo.copy()
    temp_mas2=chislo.copy()
    temp_mas3=[0]*16
    for i in range(0,3):
        temp_mas1.pop(0)
        temp_mas2.pop(15-i)
    for i in range(0,3):
        temp_mas1.append(chislo[i])
        temp_mas2.insert(0,chislo[15-i]) 
    for i in range(0,16):
        temp_mas3[i]=(temp_mas1[i]+temp_mas2[i])%10
    t=[int(''.join(map(str,temp_mas3)))]
    seed=t[0]
    return (seed/10000000000000000)
```

```python
numbers=[]
one=RNG_unlimited()
two=RNG_unlimited()
if one>two:
    min=two
    max=one
else:
    min=one
    max=two
    
print(min)
print(max)
for i in range (0,1000000):
    numbers.append(RNG_unlimited())
    if numbers[i]>max:
        max=numbers[i]
    else:
        if numbers[i]<min:
            min=numbers[i]   
print(min)
print(max)
delta=(max-min)/20

print(delta)
p_prac=[0]*21
p_prac_del=[0]*20
for i in range(0,21):
    p_prac[i]=[delta*i]
print(p_prac)
for i in range(0,1000000):
    for j in range(0,20):
        if ((numbers[i]-min)/delta>=j and (numbers[i]-min)/delta<j+1):
            p_prac_del[j]+=1
            break
p_prac_del[19]+=1
omega_teor=[0.05]*20
print(p_prac_del)
omega_prac=[0]*20
for i in range (0,20):
    omega_prac[i]=p_prac_del[i]/1000000
print(omega_prac)
```
